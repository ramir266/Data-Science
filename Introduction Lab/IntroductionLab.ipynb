## Interesting Trends and Graphs

```python
import matplotlib.pyplot as plt
import seaborn as sns

ax = sns.countplot(x="class", data=cancer).set_title("Total Count of Cancer Reocurrence")
```

Looking at this bar graph it is easily noticeable that there is more counts of no-recurrence-events and there is much less recurrance-events. 

```python
 wx = sns.countplot(x="class", hue="age", data=cancer).set_title("Total Count of Cancer Reoccurance by Age")
```

Looking at this graph, I again separated the data by the whether cancer reoccured but added another elemented. In each of there classes I separated it by age. What I find interesting is that in both classes, the most popular occurances were between the ages 40-49 and 50-59. 

```python
 fx = sns.countplot(x="tumorsize", hue="class", data=cancer).set_title("Total Count of Tumor Size by Class")
```

This is a graph of the different classes based on the tumor sizes. What I find interesting is the fact that a tumor size that is around 30-34 has the most recurrence events of cancer out of all of the other tumor sizes. 

```python
bx = sns.countplot(x="class", hue="breast-quad", data=cancer).set_title("Total Count of Class by Breast-Quad")
```

This is a graph of the reoccurance events separated by the certain breast-quad. What is interesting about this graph is the fact that both of the classes have the left_low and the left_up quad of the breast that contains cancer. 

```python
mx = sns.countplot(x="class", hue="menopause", data=cancer).set_title("Total Count of Class by Menopause")
```

This last graph is a graph of the reoccurance events which is separated by the different menopause. As you can see with both classes, the premeno and the ge40  menopause are always the dominating factors.
